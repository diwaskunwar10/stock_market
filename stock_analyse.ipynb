```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Preprocessing the dataframe
df['Date'] = pd.to_datetime(df['Date']) # Convert Date column to datetime
df = df.set_index('Date')  # Set Date column as index

# Extracting the time series of Close prices
close_series = df['Close']

# Fit a SARIMA model
model = SARIMAX(close_series, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12), maxiter=1000, method='lbfgs')

result = model.fit()

# Make predictions
forecast = result.forecast(steps=len(df))

# Evaluate the model
mse = mean_squared_error(close_series, forecast)
print("Mean Squared Error (SARIMA):", mse)
```
